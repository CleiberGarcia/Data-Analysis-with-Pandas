## Filtering Rows with the Where Method

```python
# Load the DataFrame "jamesbond.csv" indexing by "Film"
bond = pd.read_csv("jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True)
bond.head(5)
```

```python
bond["Actor"] == "Sean Connery"
```

```python
mask1 = bond["Actor"] == "Sean Connery"
```

```python
# Retrieve the rows with "Sean Connery" as Actor
bond[mask1]
```

```python
# Retrieve the rows with "Sean Connery" as Actor
bond.where(bond["Actor"] == "Sean Connery")
```

```python
# Retrieving rows with "Sean Connery" as Actor and Box Office greater than 800 (US$)
mask1 = bond["Actor"] == "Sean Connery"
mask2 = bond["Box Office"] > 800

bond.where(mask1 & mask2)
```

```python
# Retrieving rows with "Sean Connery" as Actor and Box Office greater than 800 (US$)
bond[mask1 & mask2]
```

## Retrieving Rows with the Query Method

```python
# Load the DataFrame "jamesbond.csv" indexing by "Film"
bond = pd.read_csv("jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True)
bond.head(5)
```

```python
# Replacing " " by "_" for the method query work
bond.columns = [column_name.replace(" ", "_") for column_name in bond.columns]
```

```python
bond.head(1)
```

```python
# Retrieve rows with "Sean Connery" as Actor
bond.query('Actor == "Sean Connery"')
```

```python
# Retrieve rows for which "Sean Connery" was not the JB Actor
bond.query("Actor != 'Sean Connery'")
```

```python
# Retrieve rows with "Pierce Brosnan" as JB Actor and "Michael Apted" as Director
bond.query("Actor == 'Pierce Brosnan' and Director == 'Michael Apted'")
```

```python
# Retrieve rows with "Pierce Brosnan" or "Daniel Craig" as JB Actors
bond.query("Actor in ['Pierce Brosnan', 'Daniel Craig']")
```

```python
# Retrieve rows for which "Pierce Brosnan" and "Daniel Craig" are not JB Actors
bond.query("Actor not in ['Pierce Brosnan', 'Daniel Craig']")
```
