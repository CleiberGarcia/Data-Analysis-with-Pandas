## Import Pandas and Numpy libraries

```python
import pandas as pd
import numpy as np
```

## Chainging Options with Attributes and Dot Sintax

```python
# Define an array 1000 x 50 of random integers between 0 and 100
data = np.random.randint(0, 100, [1000, 50])
data
```

```python
# Convert data to Pandas DataFrame and stores it in df variable
df = pd.DataFrame(data)
df.tail(2)
```

```python
# Show the max rows displayed by Pandas
pd.options.display.max_rows
```

```python
# Set the max rows displayed by Pandas to 10
pd.options.display.max_rows = 10
```

```python
df
```

```python
# Show the max columns displayed by Pandas
pd.options.display.max_columns
```

```python
# Set the max columns displayed by Pandas to 10
pd.options.display.max_columns = 10
```

```python
df
```

## Changing Pandas Options with Methods

```python
# Define an array 1000 x 50 of random integers between 0 and 100
data = np.random.randint(0, 100, [1000, 50])

# Convert data to Pandas DataFrame and stores it in df variable
df = pd.DataFrame(data)
df.tail(2)
```

```python
# Get the max rows number Pandas was set to display
pd.get_option("display.max_rows")
```

```python
# Get the max colums number Pandas was set to display
pd.get_option("display.max_columns")
```

```python
# Set Pandas to display maximum of 30 rows
pd.set_option("display.max_rows", 30)
```

```python
df
```

```python
# Get the max rows number Pandas was set to display
pd.get_option("display.max_rows")
```

```python
# Set Pandas to display maximum of 20 columns
pd.set_option("display.max_columns", 20)
```

```python
df
```

```python
# Get the max columns number Pandas was set to display
pd.get_option("display.max_columns")
```

```python
# Reset max rows displayed by Pandas to the original value
pd.reset_option("display.max_rows")
```

```python
# Get the max rows number Pandas was set initially to display
pd.get_option("display.max_rows")
```

```python
# Reset max columns displayed by Pandas to the original value
pd.reset_option("display.max_columns")
```

```python
# Get the max columns number Pandas was set initially to display
pd.get_option("display.max_columns")
```

```python
pd.describe_option("display.max_columns")
```

## The precision Option

```python
# Define a Pandas DataFrame of 5x5 random values
df = pd.DataFrame(np.random.randn(5,5))
df
```

```python
# Show the precision for float data type values
pd.get_option("display.precision")
```

```python
# Set precision to 2 decimal values
pd.set_option("display.precision", 2)
```

```python
df
```

```python
# Reset precision to the initial (original) value
pd.reset_option("display.precision")
```

```python
df
```
