### Converting one Data Type to Another Data Type (astype method)

```python
nba = pd.read_csv("nba.csv")
nba
```

```python
nba_df3 = nba.dropna(how="all") # deletes the last line (with all values NaN)
nba_df3.tail()
```

```python
# Note: when a column has NaN values, the conversion type may fail
nba_df3["Age"].hasnans # verify wether column "Age" has NaN values
```

```python
nba_df3["Age"]
```

```python
nba_df3.loc[:,("Age")] = nba_df3["Age"].astype("int") # convert "Age" to type int
```

```python
nba_df3["Age"]
```

### Working with Categorial Data

```python
nba_df4 = nba_df3.copy()
```

```python
nba_df4
```

```python
nba_df4.info()
```

```python
nba_df4["Position"]
```

```python
len(nba_df4) # count the number of rows
```

```python
nba_df4["Position"].nunique() # count the number of unique values
```

```python
# Convert "Position" column to datatype category
nba_df4["Position"] = nba_df4["Position"].astype("category")
```

```python
# Convert "Team" column to datatype category
nba_df4["Team"].nunique()
```

```python
# Convert "Team" column to datatype category
nba_df4["Team"] = nba_df4["Team"].astype("category")
```

### Sorting a DataFrame

#### Sorting with one column

```python
nba_df5 = nba_df4.copy()
```

```python
nba_df5["Name"]
```

```python
nba_df5["Name"].sort_values()
```

```python
nba_df5.sort_values(by = "Name", ascending = True)
nba_df5.sort_values(by = "Name")
nba_df5.sort_values("Name")
```

```python
nba_df5.sort_values("Salary", na_position = "first")
```

#### Sorting by multiple columns

```python
# Sort the nba_df5 DataFrame by Team (in ascending order) and 
# Name (in descending order)
nba_df5 = nba_df5.sort_values(by = ["Team", "Name"], ascending = [True, False])
nba_df5
```

#### Sorting a DataFrame by its index

```python
nba_df5
```

```python
nba_df5.sort_index() # sort nba_df5 DataFrame by its index
```

### Ranking values in a DataFrame

```python
import pandas as pd
```

```python
nba_df = pd.read_csv("../pandas_datasets/nba.csv")
```

```python
nba_df
```

```python
nba_df["Salary"] = nba_df["Salary"].fillna(0).astype("int")
nba_df
```

```python
nba_df["Salary Rank"] = nba_df["Salary"].rank(ascending = False).astype("int")
nba_df
```

```python
nba_df.sort_values("Salary", ascending = False)
```
